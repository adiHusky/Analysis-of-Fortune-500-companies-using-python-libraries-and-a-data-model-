```python
#1 what user posted this submission which has Raider Power Horse?
pd.read_sql_query('select * from reddit_users_posts_details where RUP_Text like "%Power Armour%"', conn)
```

```python
#2 when did user post this?
pd.read_sql_query('select RUP_Time from reddit_users_posts_details where RUP_Text=(select RUP_Text from reddit_users_posts_details where RUP_Text like "%Power Armour%")', conn)
```

```python
#3 what posts have this user posted in last 7 days?
pd.read_sql_query("select * from reddit_users_posts_details WHERE julianday('now')-julianday(RUP_Time)<julianday('now')-(julianday('now')-7) and RUP_Author_Name='KHORNEFAKES'",conn)
```

```python
#4 how many posts have this user posted in last 7 days?
pd.read_sql_query("select count(*) from reddit_users_posts_details WHERE julianday('now')-julianday(RUP_Time)<julianday('now')-(julianday('now')-7) and RUP_Author_Name='KHORNEFAKES'",conn)
```

```python
#5 what subreddits are popular?
pd.read_sql_query("select RU_Title, RU_Upvotes from Reddit_Users_Details where RU_Upvotes = (Select max(RU_Upvotes) from Reddit_Users_Details )", conn)
```

```python
#6 what posts are popular ?
pd.read_sql_query('''select  T_CEO_Text 
from Twitter_CEO_Details
where T_CEO_Retweet_Count = (Select max(T_CEO_Retweet_Count)
from Twitter_CEO_Details )''',conn)
```

```python
#7 select posts which are retweeted and company name is disney
pd.read_sql_query('select * from Twitter_CEO_Details where  T_CEO_Text like "RT%" and T_Company_Name="disney"',conn)
```

```python
#8 Select count of submissions having sector as Technology
pd.read_sql_query("""select count(RU_Submission_ID) from 
                  reddit_users_details where RU_Company_Name 
                  IN ( select  company_name from company_details_table
                  where sector='Technology')""",conn)
```

```python
#9 Title of CEO having highest number of employees
pd.read_sql_query("""SELECT ceo_name,CEO_Title from Company_CEO_Table
where company_id = (select company_id 
from company_ranking_table 
where company_employees = (select max(company_employees) 
from company_ranking_table) )""",conn)
```

```python
#10 sector of industry having lowest job growth
pd.read_sql_query('''select sector,count() cnt
                  from company_details_table as cdt2 
                  group by sector having job_growth='no' order by cnt LIMIT 1 
                 ''',conn)
```

```python
#11 Total number of employees in NJ state
pd.read_sql_query('''select sum(company_employees) from company_ranking_table where company_id in (select company_id from company_addresses_table WHERE company_HQ_state="NJ")''',conn)
```

```python
#12 Total number of woman CEOs
pd.read_sql_query('''select count(*) from company_ceo_table where ceo_woman="yes"''',conn)
```

```python
#13 company having highest revenue
pd.read_sql_query('''select company_name 
                  from  
                  company_details_table 
                  where company_id =(select company_id 
                  from company_revenue_table 
                  where company_revenue=(select max(company_revenue) 
                  from company_revenue_table))''',conn)
```
